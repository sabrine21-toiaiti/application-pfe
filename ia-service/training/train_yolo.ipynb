```python
from getpass import getpass
from roboflow import Roboflow

API_KEY = getpass('Colle ta clé API Roboflow : ')
WORKSPACE = 'steves-workspace-ls3rm'
PROJECT = 'cable-defect-4umnl'
DOWNLOAD_DIR = 'data'

rf = Roboflow(api_key=API_KEY)
project = rf.workspace(WORKSPACE).project(PROJECT)

dataset = None
last_error = None
for version_number in range(1, 21):
    try:
        print(f'Tentative de telechargement de la version {version_number}...')
        version = project.version(version_number)
        dataset = version.download('yolov8', location=DOWNLOAD_DIR)
        print(f'Dataset telecharge depuis la version {version_number}: {dataset.location}')
        break
    except Exception as error:
        last_error = error
        print(f'Version {version_number} indisponible: {error}')

if dataset is None:
    raise RuntimeError('Aucune version Roboflow telechargeable entre 1 et 20.') from last_error
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
model.train(data='data.yaml', epochs=50, imgsz=640, name='sebn_detection')
```
